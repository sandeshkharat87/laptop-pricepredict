```python
## Storage Type
df['Memory'] = df['memory'].astype(str).replace('\.0', '', regex=True)
df["Memory"] = df["memory"].str.replace('GB', '')
df["Memory"] = df["memory"].str.replace('TB', '000')
new = df["Memory"].str.split("+", n = 1, expand = True)

df["first"]= new[0]
df["first"]=df["first"].str.strip()

df["second"]= new[1]

df["Layer1HDD"] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer1SSD"] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer1Hybrid"] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer1Flash_Storage"] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['first'] = df['first'].str.replace(r'\D', '')

df["second"].fillna("0", inplace = True)

df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['second'] = df['second'].str.replace(r'\D', '')

df["first"] = df["first"].astype(int)
df["second"] = df["second"].astype(int)

df["HDD"]=(df["first"]*df["Layer1HDD"]+df["second"]*df["Layer2HDD"])
df["SSD"]=(df["first"]*df["Layer1SSD"]+df["second"]*df["Layer2SSD"])
df["Hybrid"]=(df["first"]*df["Layer1Hybrid"]+df["second"]*df["Layer2Hybrid"])
df["Flash_Storage"]=(df["first"]*df["Layer1Flash_Storage"]+df["second"]*df["Layer2Flash_Storage"])
```

```python
##preprocessing
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.model_selection import train_test_split

sc = StandardScaler()
ohe = OneHotEncoder(sparse=False, handle_unknown="ignore")  
```

```python
##PipeLine
p_cats = make_pipeline(ohe)
p_nums = make_pipeline(sc)
```

```python
##TF
TF = make_column_transformer(
(p_cats, cat_cols),
(p_nums, num_cols),    
remainder="passthrough"  )
```

```python
TF.fit_transform(X)
```

```python
## applying transformer on train data

x_train,x_test,y_train,y_test = train_test_split(X,Y, random_state=101, test_size=0.20)
```

```python
x_train.shape
```

```python
TF.fit_transform(x_train)
```

```python
from sklearn.ensemble import RandomForestRegressor,ExtraTreesRegressor,GradientBoostingRegressor,AdaBoostRegressor
rfc = RandomForestRegressor(200)
```

```python
##final Pipe
# final_pipe = make_pipeline(TF,rfc)
final_pipe = make_pipeline(TF,rfc)

```

```python
final_pipe.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
"""
# rfc 0.9038217607196969
# ete  0.8957281222346286
# ete hyper 0.8980569940512443
# ada 0.8137957158383764
# gbst 84
"""
r2_score(y_test,final_pipe.predict(x_test) )
```

```python
def pricep(sample_X, sample_Y):
    sample_X = sample_X.sample(5)
    ans = np.exp(final_pipe.predict(sample_X))
        
    y_index = sample_X.index
    y_index = list(y_index)    
    y_ans = np.exp(sample_Y.loc[y_index])
    y_ans = list(y_ans)
    
    new_df = pd.DataFrame(data=ans,columns=['pred'])
    new_df['y_ans'] = y_ans
    new_df['index'] = sample_X.index
    new_df['y_index'] = y_index
    
    return new_df

pricep(x_test,y_test)
```
